```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('/mnt/hdd0/syao/data/data_process/BGC_TD_dataset_10.csv',index_col=0)
group_dataset = dataset.groupby('isBGC')
group_dataset = group_dataset.apply(lambda x: x.sample(group_dataset.size().min()).reset_index(drop=True))
```

```python
dataset
```

```python
posdf = group_dataset[group_dataset['isBGC'] == 'Yes']
negdf = group_dataset[group_dataset['isBGC'] == 'No']
```

```python
negdf
```

```python
posdf
```

```python
neg_list = []
tmp = negdf['TDsentence'].agg(lambda x: x.split())
for i in tmp:
    neg_list+=i
neg_list
```

```python
max_len = 128
possentence_df = pd.DataFrame(posdf['TDsentence'].apply(lambda x: x.split())).reset_index(drop=True)
poslabel_df = pd.DataFrame(posdf['TDlabels'].apply(lambda x: x.split())).reset_index(drop=True)
poslabel_df
```

```python
possentence_df
```

```python

for i in range(len(possentence_df['TDsentence'])):
    random_num = np.random.randint(0,len(neg_list)-max_len)# 0~len-neg_len,é˜²æ­¢è¶…å‡ºlistè¾¹ç•Œ
    for j in range(max_len):
        if poslabel_df['TDlabels'][i][j] == '1':
            possentence_df['TDsentence'][i][j] = neg_list[random_num+j]
```
